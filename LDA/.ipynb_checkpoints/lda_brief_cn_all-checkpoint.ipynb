# LDA

## all brief paragraph

```python
cn_desc=[]
for i in range(2010,2021):
    file_path = "../LDA/segment/data/cn_file_brief_paragraph_{}.txt".format(i)
    file_cn = open(file_path, "r")
    cn_lines = file_cn.readlines()
    cn_desc_year = [i.replace(" \n","").split(" ") for i in cn_lines]
    cn_desc += cn_desc_year
    print(i,"=============")
    print(len(cn_desc_year))
    
```

```python
len(cn_desc)
```

```python
id2word = corpora.Dictionary(cn_desc)
print('資料集字典：')
print(id2word)
```

```python
corpus = [id2word.doc2bow(text) for text in cn_desc]
print('第一篇文章 corpus :')
print(corpus[0][:10])
```

+ corpus = 文檔語料庫
+ id2word = 詞彙字典
+ num_topics = 生成幾個主題數
+ random_state = 固定亂數值，每次的模型結果會一樣
+ iteration = 每個文章訓練的次數，可以設定高一點讓模型收斂
+ passes(epoch) = 整個 corpus 訓練模型的次數
+ alpha = 文章主題分佈
+ eta = 主題字分佈

```python
topic_num_list = np.arange(2, 30, 1)
result = []
for topic_num in topic_num_list:
    # perplexity
    tmp_ldamodel = gensim.models.ldamulticore.LdaMulticore(
        corpus = corpus,
        num_topics = topic_num ,
        id2word=id2word,
        random_state = 1500,
        passes=5,
        iterations=2000
        )
    tmp_perplexity = tmp_ldamodel.log_perplexity(corpus)
    temp_skleran = np.exp(-1. * tmp_perplexity)
    
    result.append({"topic_num": topic_num, "gensim_perplexity": tmp_perplexity, "perplexity":temp_skleran})
    print(topic_num)
```

```python
result = pd.DataFrame(result)
result.plot.line(x='topic_num', y='gensim_perplexity', label='gensim_perplexity')
result
```

```python
result = pd.DataFrame(result)
result.plot.line(x='topic_num', y='perplexity', label='perplexity')
result
```

```python
# 儲存LDA_result
result.to_csv("./result/LDA_cn_brief_paragraph_all.csv", index=False)
```

```python
n=0
db_cn_list=[db_rm_cn_2010,db_rm_cn_2011,db_rm_cn_2012,db_rm_cn_2013,db_rm_cn_2014,db_rm_cn_2015,db_rm_cn_2016,db_rm_cn_2017,db_rm_cn_2018,db_rm_cn_2019,db_rm_cn_2020]
topics_cn = [] 
ls_id = []
ls_pNo = []
ls_com = []
ls_desc = []
ls_date = []
ls_class = []
theta_cn = []
theta_all_cn = []


for k in range(2010,2021):
    f_path = "../LDA/segment/data/cn_file_brief_paragraph_{}.txt".format(k)
    file_cn = open(f_path, "r")
    cn_lines = file_cn.readlines()
    cn_desc = [i.replace(" \n","").split(" ") for i in cn_lines]

    # 取得每個文章的主題分佈
    id2word = corpora.Dictionary(cn_desc)
    corpus = [id2word.doc2bow(text) for text in cn_desc]
    topics_doc = ldamodel.get_document_topics(corpus, minimum_probability=0.000001)
    
    # topics_cn = [] 
    for topics in topics_doc:
        c = [i[1] for i in topics]
        topics_cn.append(c.index(max(c)))
        theta_cn.append(max(c))
        theta_all_cn.append(c)


    # 找出對應的patent編號和公司
    # for i in db_cn_list[n]:
    data_cn_desc = db_cn_list[n].find({}, {"patentNumber":1, "brief": 1, "current_assignee":1, "appDate":1, "classification":1})

    # ls_id = []
    # ls_pNo = []
    # ls_com = []
    # ls_desc = []
    for data in data_cn_desc:
        ls_id.append(data["_id"])
        ls_pNo.append(data["patentNumber"])
        ls_com.append(data["current_assignee"])
        ls_desc.append(data["brief"])
        ls_date.append(data["appDate"])
        ls_class.append(data["classification"])

    

    print(k,"===============")
    print(len(topics_cn))
    print(len(ls_desc))

    n+=1
```

```python
df_cn = pd.DataFrame((zip(ls_id, ls_pNo, ls_date, ls_com, ls_class, ls_desc)), columns = ['id', 'patentNumber', 'appDate', 'company', 'classification', 'brief'])
df_cn["topic"]=topics_cn
df_cn["theta"]=theta_cn
df_cn["theta_all"]=theta_all_cn
```
