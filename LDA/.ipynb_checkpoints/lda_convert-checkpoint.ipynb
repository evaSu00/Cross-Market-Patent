## Forbes list  
- 同Industry：1
- 同Sector：2
- 有出現在company list：3
- Company list上有但Forbes沒有：4

```python
f_data = pd.read_csv("../patent/code/依年分/Forbes/Forbes_2020.csv")
```

```python
forb_ls =[]

for i in com_ls:
    if (f_data["Company"]==i).any():
        if (f_data[f_data.Company==i]["Industry"]=="Semiconductors").any():
            forb_ls.append(1)
        elif (f_data[f_data.Company==i]["Sector"]=="Information Technology").any():
            forb_ls.append(2)
        else:
            forb_ls.append(3)
    else:
        forb_ls.append(4)
    
```

```python
forb_ls
```

```python
lda_all["forb_2020"] = forb_ls
```

```python
lda_all
```
