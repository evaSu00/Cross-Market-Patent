## 找出topic最多的CPC

```python
df_cn_all_n.groupby(["topic","cpc_n"],as_index=False).size()
```

```python
df_cn_all_s = df_cn_all_n.groupby(["topic","cpc_n"],as_index=False).size()
df_cn_all_s
```

```python
df_cn_all_s = df_cn_all_s[(df_cn_all_s["cpc_n"]!="H04")]
```

```python
df_cn_all_s[df_cn_all_s["topic"]==8].sort_values("size",ascending=False).head(10)
```

```python
df_cn_all_s[df_cn_all_s["topic"]==5].sort_values("size",ascending=False).head(10)
```

```python
cpc_max = df_cn_all_s.groupby(["topic"],as_index=False)["size"].agg('max')
cpc_max
```

```python
df_cn_all_s.merge(cpc_max,on = ["topic","size"])
```

```python
# 轉成 Year-topic 的總數表格
table_cn = df_cn_all_s.pivot_table(index='topic', values='size', columns='cpc_n',aggfunc=['sum']).fillna(0).reset_index()
table_cn = table_cn.droplevel(level=0, axis=1)
table_cn = table_cn.rename(columns={'':'topic'})
```

```python
table_cn
```

```python
table_cn.to_csv('lda_all_result/lda_brief_cn_all_topic_cpc_11.csv',index=0)
```
