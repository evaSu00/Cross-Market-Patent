# Data

```python
def get_forb(f_data,com_ls):
    forb_ls =[]

    for i in com_ls:
        try:
            if (f_data["Company"]==i).any():
                if (f_data[f_data.Company==i]["Industry"]=="Semiconductors").any():
                    forb_ls.append(1)
                elif (f_data[f_data.Company==i]["Sector"]=="Information Technology").any():
                        forb_ls.append(2)
                else:
                    forb_ls.append(3)
            else:
                forb_ls.append(4)
        except:
            if (f_data["Company"]==i).any():
                if (f_data[f_data.Company==i]["Industry"]=="Semiconductors").any():
                    forb_ls.append(1)
                # elif (f_data[f_data.Company==i]["Sector"]=="Information Technology").any():
                #         forb_ls.append(2)
                else:
                    forb_ls.append(3)
            else:
                forb_ls.append(4)
    
    return forb_ls
```

## US

```python
nmi_all=[]
ari_all=[]
for i in range(2015,2021):
    now = "../competitor_df/lda_us_{year}.csv".format(year=i)
    cpc_now = pd.read_csv(now)
    
    nmi_ls=[]
    ari_ls=[]
    print(i)
    
    for j in range(2010,2021):
        f = "../../patent/code/依年分/Forbes/Forbes_{year}.csv".format(year=j)
        f_data = pd.read_csv(f)

        com_ls = list(cpc_now.company)

        forb_ls = get_forb(f_data,com_ls)
        competitor_ls = list(cpc_now.iloc[:,1])

        # cpc_now["forb"] = forb_ls
        
        nmi = normalized_mutual_info_score(forb_ls,competitor_ls)
        nmi_ls.append(nmi)

        ari = adjusted_rand_score(forb_ls,competitor_ls)
        ari_ls.append(ari)
        
        print("forb:",j)
        print("forb_ls",forb_ls)
        print("competitor_ls",competitor_ls)
        
        
    nmi_all.append(nmi_ls)
    ari_all.append(ari_ls)
    print("===============================")
```

```python
nmi_all
```

```python
ari_all
```

```python
col_ls=list(range(2010,2021))
index_ls=list(range(2015,2021))
result = pd.DataFrame(nmi_all,columns=col_ls,index=index_ls)
result
```

```python
result.to_csv("lda_us_all_year.csv",index=True)
```
