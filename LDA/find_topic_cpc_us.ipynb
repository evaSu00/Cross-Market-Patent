## 找出topic最多的CPC

```python
df_us_all_n.groupby(["topic","cpc_n"],as_index=False).size()
```

```python
df_us_all_s = df_us_all_n.groupby(["topic","cpc_n"],as_index=False).size()
df_us_all_s
```

```python
df_us_all_s = df_us_all_s[(df_us_all_s["cpc_n"]!="H04")&(df_us_all_s["cpc_n"]!="G06")]
```

```python
df_us_all_s[df_us_all_s["topic"]==7].sort_values("size",ascending=False).head(10)
```

```python
df_us_all_s[df_us_all_s["topic"]==1].sort_values("size",ascending=False).head(10)
```

```python
cpc_max = df_us_all_s.groupby(["topic"],as_index=False)["size"].agg('max')
cpc_max
```

```python
df_us_all_s.merge(cpc_max,on = ["topic","size"])
```

```python
# 轉成 Year-topic 的總數表格
table_us = df_us_all_s.pivot_table(index='topic', values='size', columns='cpc_n',aggfunc=['sum']).fillna(0).reset_index()
table_us = table_us.droplevel(level=0, axis=1)
table_us = table_us.rename(columns={'':'topic'})
```
